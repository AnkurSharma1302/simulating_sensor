```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
sensor_data = pd.read_csv('./temperature_readings.csv', encoding='latin1')
```

```python
sensor_data.head()
```

```python
def plot_temperature_data(sensor_data):
    # Extract data for plotting
    minutes = sensor_data["Minute"]
    temperatures = sensor_data["Temperature (°C)"]
    
    # Create the plot
    plt.figure(figsize=(12, 6))
    plt.plot(minutes, temperatures, label="Temperature (°C)", color="blue", linewidth=2)
    
    # Highlight overheating values
    overheating_minutes = minutes[temperatures > 80]
    overheating_temperatures = temperatures[temperatures > 80]
    plt.scatter(overheating_minutes, overheating_temperatures, color="red", label="Overheating (> 80°C)", zorder=5)
    
    # Add a threshold line
    plt.axhline(y=80, color="orange", linestyle="--", linewidth=1, label="Overheating Threshold (80°C)")

    # Add labels, title, and legend
    plt.title("Washing Machine Temperature During Wash Cycle", fontsize=16)
    plt.xlabel("Time (Minutes)", fontsize=12)
    plt.ylabel("Temperature (°C)", fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(alpha=0.5)

    # Show the plot
    plt.show()

# Plot the data
plot_temperature_data(sensor_data)
```
